```python
import numpy as np
import pandas as pd
import difflib as dfl
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data_1=pd.read_csv("C:\\Users\\HP\\Downloads\\song_data.csv")
data_1.head(50)
```

```python
data_2=pd.read_csv("C:\\Users\\HP\\Downloads\\triplets_file.csv")
data_2.head(50)
```

```python
data_3 = pd.merge(data_2, data_1.drop_duplicates(['song_id']), on='song_id', how='left')
display (data_3.shape)
display (data_3.head(50))
```

```python
data_3=data_3.head(60000)
data_3.shape
```

```python
data_3['song'] = data_3['title']+' - '+data_3['artist_name']
data_3.head()
```

```python
song_grouped = data_3.groupby(['song']).agg({'listen_count':'count'}).reset_index()
song_grouped.head()
```

```python
grouped_sum = song_grouped['listen_count'].sum()
grouped_sum
```

```python
song_grouped['percentage'] = (song_grouped['listen_count'] / grouped_sum ) * 100
song_grouped.sort_values(['listen_count', 'song'], ascending=[0,1])
```

```python
combined_features = song_grouped['song']+' '+str(song_grouped['listen_count'])+' '+str(song_grouped['percentage'])
display (combined_features)
```

```python
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
display (feature_vectors.shape)
print (feature_vectors)
```

```python
similarity = cosine_similarity(feature_vectors)
print  (similarity )
```

```python
similarity.shape
```

```python
song_name = input(' Enter song name with its artist: ')
```

```python
all_titles = list(data_3['title'])
print(all_titles)
```

```python

find_close_match = dfl.get_close_matches(song_name,all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
song_index = data_3[data_3.title == close_match]['year'].values[0]

similarity_score = list(enumerate(similarity[song_index]))

similar_songs = sorted(similarity_score, key = lambda x:x[1], reverse = True) 

print('Songs suggested for you : \n')

i = 1

for song in similar_songs:
  year = song[0]
  title_from_index = data_3[data_3.index==year]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
